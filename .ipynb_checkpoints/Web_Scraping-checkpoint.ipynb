```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
page = requests.get("https://www.iitkgp.ac.in/department/ME/faculty/me-a.guha")
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
divs = soup.find_all('div', class_='accordion-body')
list = []
```

```python
for i in divs:
    item = [i.text.strip()]
    list = list+item
```

```python
print(list)
```

```python
link = soup.find_all('a', class_="accordion-anch")
links = []
for i in link:
    links.append(i.get('href'))
```

```python
mails = []
```

```python
for i in links:
    page = requests.get(i)
    soup = BeautifulSoup(page.text, 'html')
    mails.append(soup.find('p', class_="tx01 me-5 mb-0").text.strip())
```

```python
position = []
```

```python
for i in links:
    page = requests.get(i)
    soup = BeautifulSoup(page.text, 'html')
    soup1 = soup.find('div', class_="con_bk")
    position.append(soup1.find('i').text.strip())
```

```python
Mechanical_Professor_Details = pd.DataFrame({'Name': list ,'Position':position, "Mail-Id":mails, "Web-Page":links})
```
