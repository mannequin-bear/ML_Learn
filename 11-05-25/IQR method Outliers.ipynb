```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('placement.csv')
df.head()
```

```python
plt.figure(figsize=(16,10))
df['placement_exam_marks'].plot(kind="box")
```

```python
df['cgpa'].skew()
```

```python
df['placement_exam_marks'].skew()
```

```python
# finding the iqr

q1 = df['placement_exam_marks'].quantile(0.25)
q3 = df['placement_exam_marks'].quantile(0.75)
iqr = q3-q1
upper_limit = q3 + (1.5*iqr)
lower_limit = q1 - (1.5*iqr)
```

```python
df[df['placement_exam_marks'] > upper_limit]
```

### trim

```python
new_df = df[df['placement_exam_marks'] < upper_limit] 
```

```python
new_df['placement_exam_marks'].plot(kind='box')
```

### capping

```python
df['placement_exam_marks'] = np.where(df['placement_exam_marks'] > upper_limit, upper_limit, np.where(df['placement_exam_marks'] < lower_limit, lower_limit, df['placement_exam_marks']))
```

```python
df['placement_exam_marks'].plot(kind='box')
```
