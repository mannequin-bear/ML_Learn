```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('weight-height.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df['Height'].plot(kind="kde")
```

```python
df['Height'].plot(kind="box")
```

```python
upper_limit = df['Height'].quantile(0.99)
lower_limit = df['Height'].quantile(0.01)
upper_limit, lower_limit
```

### Trimming 


```python
new_df = df[(df['Height'] <= 74.78) & (df['Height'] >= 58.13)]
```

```python
new_df['Height'].plot(kind='kde')
```

```python
new_df['Height'].plot(kind='box')
```

### Winsorization / Capping

```python
df_copy = df
```

```python
df_copy['Height'] = np.where(df_copy['Height'] >= upper_limit, upper_limit, np.where(df_copy['Height'] <= lower_limit, lower_limit, df_copy['Height']))
```

```python
df_copy['Height'].plot(kind='box')
```
